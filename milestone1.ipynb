## Step 2

Import essential Python libraries for data analysis and visualization.

```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib import cm
```

## Step 3

To ensure accurate analysis, we must convert relevant columns to numerical values and remove missing data. 

So, we go through each dataset and convert any strings & non-numerical values to numerical values to use for graphing and plotting. Also, ensure all unnecessary white spaces and "NaN" values are eliminated as they are not useful for plotting.

```python
# Removes leading/trailing spaces for both datasets
df_surface.columns = df_surface.columns.str.strip() 
df_rain.columns = df_rain.columns.str.strip()  

# Converts lat & lon columns to numeric values for both datasets
df_surface['Lat'] = pd.to_numeric(df_surface.iloc[:, 8], errors='coerce') 
df_surface['Lon'] = pd.to_numeric(df_surface.iloc[:, 9], errors='coerce')
df_rain['Lat'] = pd.to_numeric(df_rain.iloc[:, 5], errors='coerce')
df_rain['Lon'] = pd.to_numeric(df_rain.iloc[:, 6], errors='coerce')

# Drops NaN values from lat & lon columns for both datasets
df_surface_clean = df_surface.dropna(subset=['Lat', 'Lon']) 
df_rain_clean = df_rain.dropna(subset=['Lat', 'Lon'])

# Convert to numeric values and remove NaN values for pressure, gust, and storm surge columns in the surface data set
df_surface_clean.loc[:, 'Pressure (mb)'] = pd.to_numeric(df_surface_clean['Pressure (mb)'], errors='coerce')
df_surface_clean1 = df_surface_clean.dropna(subset=['Pressure (mb)'])
df_surface_clean.loc[:, 'Gust (kt)'] = pd.to_numeric(df_surface_clean['Gust (kt)'], errors='coerce')
df_surface_clean2 = df_surface_clean.dropna(subset=['Gust (kt)'])
df_surface_clean.loc[:, 'Storm Surge (ft)'] = pd.to_numeric(df_surface_clean['Storm Surge (ft)'], errors='coerce')
df_surface_clean3 = df_surface_clean.dropna(subset=['Storm Surge (ft)'])

# Remove NaNs after conversion of total rainfall column to numeric values for rainfall dataset
df_rain_clean.loc[:, 'Totals (inches)'] = pd.to_numeric(df_rain_clean['Totals (inches)'], errors='coerce')
df_rain_clean = df_rain_clean.dropna(subset=['Totals (inches)'])
```

## Step 4

Plot the minimum sea level pressure of Hurricane Ian as it made landfall in Florida with scatterplots showing where the minimum sea level pressure of the storm was located as it made landfall. 

This is done to analyze the storm's movement and intensity, as well as to visualize areas of Florida that experience more severe impacts from the storm, as areas with lower sea level pressure likely experienced the storms' initial landfall and the initial devastating impacts. 

The plot shows different colored scatter points over Florida, with warmer colors representing areas of higher pressure and cooler colors representing areas of lower pressure. We can see from the plot that areas in SW Florida experienced much lower pressure compared to other areas of the state, indicating that SW Florida was where the storm made landfall.

```python
def plot_hurricane_track(df_surface_clean):  
    # Drop NaN values
    df_surface_clean = df_surface_clean.dropna(subset=['Lat (°)', 'Lon (°)', 'Pressure (mb)'])

    # Check if the longitude values are in the wrong format (positive instead of negative for Florida)
    if df_surface_clean['Lon (°)'].min() > 0:
        df_surface_clean.loc[:, 'Lon (°)'] = -df_surface_clean['Lon (°)']  # Convert longitudes to negative

    # Filter data for Florida (lat: 24.5 to 31, lon: -87.5 to -79.5)
    df_surface_clean = df_surface_clean[(df_surface_clean['Lat (°)'] >= 24.5) & (df_surface_clean['Lat (°)'] <= 31) &
                                        (df_surface_clean['Lon (°)'] >= -87.5) & (df_surface_clean['Lon (°)'] <= -79.5)]

    # Extract values for plotting
    lons = df_surface_clean['Lon (°)'].values
    lats = df_surface_clean['Lat (°)'].values
    pressures = df_surface_clean['Pressure (mb)'].values

    # Normalize pressure for colormap
    norm = plt.Normalize(pressures.min(), pressures.max())
    cmap = plt.get_cmap('coolwarm')

    # Set up the map
    fig, ax = plt.subplots(figsize=(10, 15), subplot_kw={'projection': ccrs.PlateCarree()})

    # Set Florida map extent
    ax.set_extent([-87.5, -79.5, 24.5, 31], crs=ccrs.PlateCarree())

    # Add map features
    ax.add_feature(cfeature.LAND, color='lightgray')
    ax.add_feature(cfeature.OCEAN, color='lightblue')
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    ax.add_feature(cfeature.STATES, linestyle='--')

    # Plot the hurricane track as a scatter plot with large points and low transparency
    sc = ax.scatter(lons, lats, c=pressures, cmap=cmap, norm=norm, 
                    edgecolors='black', s=150, alpha=0.7, transform=ccrs.PlateCarree())

    # Add colorbar
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    cbar = plt.colorbar(sm, ax=ax, orientation='vertical', fraction=0.02, pad=0.04)
    cbar.set_label('Minimum Sea Level Pressure (hPa)')

    # Title
    plt.title('Minimum Sea Level Pressure Measurements Over Florida During Hurricane Ian')

    # Show the plot
    plt.show()

plot_hurricane_track(df_surface_clean)
```

## Step 5

This map displays the average total rainfall accumulation during the hurricane's passage over different weather stations Florida. The x-axis shows each date from the rainfall dataset and the y-axis is in inches. 

What this graph shows is the average rainfall at each station in Florida over each day during Hurricane Ian's landfall. This is to visualize what days had the most average rainfall, which helps us understand on what day Florida likely saw the most impacts from the storm. From this plot, we see that September 29, 2022, is when Florida experienced the highest average rainfall, with about 1.4 inches in rain.

```python
# Filter for Florida
df_rain_fl = df_rain[df_rain['State/Province'] == 'FL']

# Drop rows where 'Latitude (°)', 'Longitude (°)', or 'Totals (inches)' are NaN
df_rain_fl = df_rain_fl.dropna(subset=['Latitude (°)', 'Longitude (°)', 'Totals (inches)'])

# Filter to keep only valid latitude and longitude ranges for Florida
df_rain_fl = df_rain_fl[(df_rain_fl['Latitude (°)'] >= 24.5) & (df_rain_fl['Latitude (°)'] <= 31)]
df_rain_fl = df_rain_fl[(df_rain_fl['Longitude (°)'] >= -87.5) & (df_rain_fl['Longitude (°)'] <= -79.5)]

# Identify the date columns dynamically (columns representing daily rainfall)
date_columns = [col for col in df_rain_fl.columns if '-' in col]

# Convert the date columns to numeric values and fill NaNs with 0
df_rain_fl[date_columns] = df_rain_fl[date_columns].apply(pd.to_numeric, errors='coerce').fillna(0)

# Compute the average rainfall across all stations per day
avg_rainfall_per_day = df_rain_fl[date_columns].mean()

# Plot the average daily rainfall
plt.figure(figsize=(10, 6))
plt.plot(avg_rainfall_per_day.index, avg_rainfall_per_day.values, marker='o', color='b', linestyle='-')

# Set labels and title
plt.title('Average Daily Rainfall Across Florida Weather Stations During Hurricane Ian', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Rainfall (inches)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

## Step 6

This plot represents maximum wind gusts observed during Hurricane Ian's passage over Florida. This is a hexbin heatmap that displays longitude on the x-axis and latitude on the y-axis. The darker colors on the heatmap indicate areas of Florida that experienced lower wind gusts and brighter colors indicate areas that experienced higher wind gusts. 

This map helps visualize what areas of Florida experienced the strongest wind gusts during the storm, which helps us understand areas that had more devastating impacts from the storm. We can see from the plot that one area in around 26.5°N 82°W (likely around Cape Coral, FL) experienced maximum wind gusts close to 120 kt, indicating severe impacts from the storm at that location.

```python
# Drop NaN values for Lat (°), Lon (°), and Gust (kt) before filtering
df_surface_clean = df_surface_clean.dropna(subset=['Lat (°)', 'Lon (°)', 'Gust (kt)'])

# Filter the data for Florida 
df_surface_fl = df_surface_clean[df_surface_clean['State/Province'] == 'FL']

# Ensure longitudes are negative for Florida
df_surface_fl.loc[:, 'Lon (°)'] = df_surface_fl['Lon (°)'].apply(lambda x: -abs(x))

# Convert the latitudes, longitudes, and gusts to numpy arrays for further processing
lats = df_surface_fl['Lat (°)'].values
lons = df_surface_fl['Lon (°)'].values
gusts = df_surface_fl['Gust (kt)'].values

# Create the heatmap without a background map
plt.figure(figsize=(10, 6))

# Create a scatter plot to visualize the gusts on a heatmap
plt.hexbin(lons, lats, C=gusts, gridsize=40, cmap='plasma', reduce_C_function=np.max)

# Add a colorbar to represent gusts
plt.colorbar(label='Max Wind Gusts (kt)')

# Set the axis labels
plt.xlabel('Longitude (°)')
plt.ylabel('Latitude (°)')

# Set a title for the plot
plt.title('Maximum Wind Gusts Observed in Florida During Hurricane Ian')

# Show the plot
plt.show()
```

## Step 7

This plot contains data showing areas that experienced storm surge during Hurricane Ian's passage over Florida. It is a scatterplot that displays longitude on the x-axis and latitude on the y-axis. The darker colors on the plot indicate areas of Florida that experienced lower storm surge and brighter colors indicate areas that experienced higher storm surge. 

This map helps visualize what areas of Florida experienced the highest storm surge during the storm, which helps us understand areas that had more devastating storm surge impacts from the storm. We can see from the plot that one area by 26.5°N 82°W (around Cape Coral, FL), experienced approximately 7 ft in storm surge.

```python
# Filter the data for Florida
df_surface_fl = df_surface_clean[df_surface_clean['State/Province'] == 'FL']
df_surface_fl = df_surface_fl.dropna(subset=['Lat (°)', 'Lon (°)', 'Storm Surge (ft)'])

# Ensure longitudes are negative for Florida
df_surface_fl.loc[:, 'Lon (°)'] = df_surface_fl['Lon (°)'].apply(lambda x: -abs(x))

# Convert to numpy arrays for latitudes, longitudes, and storm surge values
lats = df_surface_fl['Lat (°)'].values
lons = df_surface_fl['Lon (°)'].values
surges = df_surface_fl['Storm Surge (ft)'].values

# Create a map without background using a scatter plot approach
plt.figure(figsize=(10, 6))

# Create a scatter plot for the storm surge values
plt.scatter(lons, lats, c=surges, cmap='viridis', s=50, edgecolors='k', alpha=0.7)

# Add colorbar to represent storm surge levels
plt.colorbar(label='Storm Surge (ft)')

# Set axis labels
plt.xlabel('Longitude (°)')
plt.ylabel('Latitude (°)')

# Set title
plt.title('Storm Surge Observations in Florida During Hurricane Ian')

# Show plot
plt.show()
```
